```python
import mxnet as mx
import numpy as np
import caffe as cf
from caffe import layers as L
import os,cv2
from mxnet2caffe import cvt_end2end

from matplotlib import pyplot as plt
%pylab inline

def show_img(img):
    fig = plt.imshow(img.clip(0,255))
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    plt.show()
def show_rgb(img,ifswap=True):
    img = np.transpose(img,axes=(1,2,0)).clip(0,255).astype(np.uint8)
    if ifswap:
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    fig = plt.imshow(img)
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    plt.show()
def show_gray(img):
    fig = plt.imshow(img)
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    plt.colorbar()
    plt.show()
def easyshow(src,title,ifcolorbar=True,cmap='gray'):
    gci=plt.imshow(src,cmap=cmap)
    plt.title(title)
    if ifcolorbar:
        plt.colorbar(gci)
    plt.show()
    return
```

## Begin inference test

```python
# prepare data
def read_nv21(srcfn,w,h):
    raw = np.fromfile(srcfn, dtype=np.uint8)
    y = raw[:w*h].reshape(h,w)
    uv = raw[w*h:]
    uv = uv.reshape((w*h/4, 2))
    u = uv[:,1].reshape(h/2,w/2)
    v = uv[:,0].reshape(h/2,w/2)
    return y,u,v

def write_nv21(outprefix, y,u,v):
    uv = np.stack((v,u),axis=-1).flatten()
    out_arr = np.concatenate((y.flatten(),uv),axis=0)
    out_arr=out_arr.astype(np.uint8)
    out_arr.tofile(outprefix+'.NV21')
    return
```

```python
src_y,src_u,src_v = read_nv21('./data-example/src-1080p_1920x1080.NV21',1920,1080)

transfer_net.blobs['data0'].data[...] = src_y
transfer_net.blobs['data1'].data[...] = src_u
transfer_net.blobs['data2'].data[...] = src_v
transfer_net.forward()
dst_y = transfer_net.blobs['ms-0070'].data[0]
dst_u = transfer_net.blobs['ms-0073'].data[0]
dst_v = transfer_net.blobs['ms-0076'].data[0]

write_nv21('test-out',dst_y,dst_u,dst_v)
```
